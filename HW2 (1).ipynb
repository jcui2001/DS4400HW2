```python
def closed_form_sol(df_, y_):
    '''
    Produces the theta weight of trained data using closed form solution
    
    df_ (DataFrame): dataframe of training data
    y_ (list): list of points of the feature to be predicted
    x_samp (list): list of values for each feature
    
    Return:
    (np.array): array of trained weights (theta) for each feature in dataframe
    
    '''
    
    # turns given df to numpy array
    matrix = df_
    # turns given y_pred list to numpy array
    y_mat = y_
    
    # transposes the df_matrix
    matrix_t = np.transpose(df_)
    
    # performs close form operations to produce list of thetas
    cfm1 = np.linalg.pinv(np.matmul(df_, matrix_t))
    cfm2 = np.matmul(matrix_t, cfm1)
    cfm3 = np.matmul(cfm2, y_)
    
    return cfm3
```

```python
def poly_reg(x_feat, y_feat, p):
    '''
    Trains polynomial regression on a given training model and test point
    
    train_df (pd.DataFrame): training data DataFrame
    theta_list (list): list of predicted theta of features
    feat (list): feature class from training data
    y_feat (str): feature class to predict
    p (int): degree of polynomial model
    
    Return:
    list: list of predicted values from test point on training data
    '''
    # range of given p
    p_range = range(1, p)
    # initialize dataframe for results
    poly_x = pd.DataFrame()
    
    # range for given p to loop
    for i in range(1, p):
        # raises every x by this p
        x_raised = [ x ** i for x in x_feat ]
        # appends x_raised to dataframe
        poly_x[f'{i}'] = x_raised
        
    result = np.matmul(poly_x, closed_form_sol(poly_x, y_feat)).to_numpy()
    
    return result
```

```python
def gradient_descent(theta, y_, x_, iters, alpha):
    '''
    Returns the adjusted theta after gradient descent is applied
    
    theta (array): matrix of theta weights derived from model
    y_ (array): 1-d array of response feature
    x_ (array): training dataset
    iters (num): number of iterations
    
    Returns:
    (array): new theta weights after GD update
    '''
    
    theta_prime = theta
    
    for ith in range(1, iters):
        
        error_term = np.subtract(np.matmul(x_, theta), y_)
        
        trans_term = np.transpose(error_term).tolist()
    
        sub_array = np.matmul(trans_term, x_)
        
        s_array = np.dot(2, sub_array)
        
        su_array = np.divide(s_array, x_.size)
    
        next_theta = np.dot(su_array, alpha)
        
        theta_prime = np.add(theta_prime, next_theta)
        
    results = np.matmul(x_, theta_prime)
        
    #return s_array
    #return theta_prime
    #return error_term
    #return trans_term
    return results, theta_prime
```

```python
def append_ridge(theta, y_, x_, iters, alpha, lam):
    '''
    Appends ridge regression term to given theta list 
    
    theta (list): 1-d array of theta values
    y_ (list): 1-d array of y to fit gradient descent
    x_ (array): 2-d array of training data for gradient descent
    iters (int): number of iterations of gradient descent
    alpha (float): alpha factor
    lam (float): lambda control factor
    
    Return:
    (array): updated training dataset with adjusted weights
    
    '''
    # uses gradient descent to produce array of updated dataset and thetas
    new_vals, adj_thetas = gradient_descent(theta, y_, x_, iters, alpha)
    
    # ridge value
    ridge_ = alpha * lam
    # multiplies ridge by each theta
    ridge = np.dot(adj_thetas, ridge_)
    
    ridge_appd = np.add(adj_thetas, ridge)
    
    x_scaled = np.reshape(x_, (-1, len(x_[0])))
    
    rdg_result = np.matmul(x_scaled, ridge_appd)
    '''
    result_scaled = np.reshape(rdg_result, (-1, 18))
    '''
    return rdg_result
    #return result_scaled
    #return x_scaled
```
