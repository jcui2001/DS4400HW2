```python
def closed_form_sol(df_, y_):
    '''
    Produces the theta weight of trained data using closed form solution
    
    df_ (DataFrame): dataframe of training data
    y_ (list): list of points of the feature to be predicted
    x_samp (list): list of values for each feature
    
    Return:
    (np.array): array of trained weights (theta) for each feature in dataframe
    
    '''
    
    # turns given df to numpy array
    matrix = df_
    # turns given y_pred list to numpy array
    y_mat = y_
    
    # transposes the df_matrix
    matrix_t = np.transpose(df_)
    
    # performs close form operations to produce list of thetas
    cfm1 = np.linalg.pinv(np.matmul(df_, matrix_t))
    cfm2 = np.matmul(matrix_t, cfm1)
    cfm3 = np.matmul(cfm2, y_)
    
    return cfm3
```

```python
def gradient_descent(theta, y_, x_, iters, alpha):
    '''
    Returns the adjusted theta after gradient descent is applied
    
    theta (array): matrix of theta weights derived from model
    y_ (array): 1-d array of response feature
    x_ (array): training dataset
    iters (num): number of iterations
    
    Returns:
    (array): new theta weights after GD update
    '''
    
    theta_prime = theta
    
    for ith in range(1, iters):
        
        error_term = np.subtract(np.matmul(x_, theta), y_)
        
        trans_term = np.transpose(error_term).tolist()
    
        sub_array = np.matmul(trans_term, x_)
        
        s_array = np.dot(2, sub_array)
        
        su_array = np.divide(s_array, x_.size)
    
        next_theta = np.dot(su_array, alpha)
        
        theta_prime = np.add(theta_prime, next_theta)
        
    #return s_array
    return theta_prime
    #return error_term
    #return trans_term
```

```python
def append_ridge(theta, y_, x_, iters, alpha, lam):
    
    new_vals = gradient_descent(theta, y_, x_, iters, alpha)
    
    ridge_ = 2 * alpha * lam
    ridge = np.dot(theta, ridge_)
    ridge_appd = np.subtract(new_vals, ridge)
    rdg_result = np.matmul(x_, ridge_appd)
    
    return rdg_result
```
